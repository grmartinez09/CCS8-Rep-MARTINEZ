```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target

from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

DBSCAN Clustering

```python
# Applying DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters_dbscan = dbscan.fit_predict(data.iloc[:, :-1])

# Plotting
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=clusters_dbscan, cmap='rainbow')
plt.title('DBSCAN Clustering')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

KMEANS Clustering

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target

from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Applying K-means
kmeans = KMeans(n_clusters=3, random_state=42)
clusters_kmeans = kmeans.fit_predict(data.iloc[:, :-1])

# Plotting
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=clusters_kmeans, cmap='rainbow')
plt.title('K-means Clustering')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

HIERARCHIAL CLUSTERING

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target

from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Applying Hierarchical Clustering
hier_clust = AgglomerativeClustering(n_clusters=3)
clusters_hierarchical = hier_clust.fit_predict(data.iloc[:, :-1])

# Plotting
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=clusters_hierarchical, cmap='rainbow')
plt.title('Hierarchical Clustering')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()

# Dendrogram
linked = linkage(data.iloc[:, :-1], method='ward')
dendrogram(linked)
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```
